```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
mpl.rcParams.update(mpl.rcParamsDefault)

%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
plt.style.use("seaborn-darkgrid")
```

```python
#PAÄ°RPLOT
fish = pd.read_csv("data/Fish.csv")
fish.head()
```

```python
plt.figure(figsize = (11,9))
sns.pairplot(fish, hue = "Species")
plt.show()
```

```python
fish1 = fish[fish["Species"].isin(["Bream","Perch","Pike"])]
plt.figure(figsize = (11,9))
sns.pairplot(fish1, hue = "Species")
plt.show()
```

```python
sns.pairplot(fish1, hue = "Species", palette = "husl", height =2)
plt.show()
```

```python
sns.pairplot(fish1,  hue = "Species" , vars = ["Length1","Weight","Height","Width"],
            markers = ["o","D","<"], height = 3, aspect =1)
plt.show()
```

```python
sns.pairplot(fish1,  hue = "Species" , vars = ["Weight","Height","Width"],
            markers = ["o","D","<"], height = 5, aspect =1)
plt.show()
```

```python
sns.pairplot(fish1,  hue = "Species" , vars = ["Weight","Height","Width"],
            markers = ["o","D","<"], height = 5, aspect =1)
plt.show()
```

```python
sns.pairplot(fish1,  hue = "Species" , vars = ["Weight","Height","Width"], kind = "reg",height = 5, aspect =1)
plt.show()
```

```python
sns.pairplot(fish1,  hue = "Species" , vars = ["Weight","Height","Width"],plot_kws = dict(s =140),height = 5, aspect =1)
plt.show()
```

```python
sns.pairplot(fish1,  hue = "Species" , 
             x_vars = ["Weight","Height"],
             y_vars = ["Weight","Height"],
             plot_kws = dict(s =140, linewidth =1,alpha = 1),height = 5, aspect =1)
plt.show()
```

```python
#PAÄ°RGRÄ°D
```

```python
g = sns.PairGrid(fish1, hue="Species")
g.map(plt.scatter)
plt.show()
```

```python
g = sns.PairGrid(fish1)
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist)
g= g.add_legend()
plt.show()
```

```python
g = sns.PairGrid(fish1, hue="Species")
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist, histtype = "step",linewidth =2)
g= g.add_legend()
plt.show()
```

```python
g = sns.PairGrid(fish1, hue="Species", vars= ["Weight","Height","Width"], height = 5, aspect=1)
g = g.map_offdiag(plt.scatter, edgecolor = "m",s=130)
g = g.map_diag(plt.hist, edgecolor = "b",linewidth =2)
g= g.add_legend()
plt.show()
```

```python
insurance = pd.read_csv("data/insurance.csv")
insurance.head()
```

```python
sns.lmplot(x = "bmi",y="expenses", hue = "smoker", data = insurance, height =8,aspect=1)
plt.show()
```

```python
sns.lmplot(x = "bmi",y="expenses", hue = "smoker", data = insurance, height =8,aspect=1,
          scatter_kws = {"s":90, "alpha": .4}, palette =["#ff3333","#00ac00"])
plt.show()
```

```python
sns.lmplot(x = "bmi",y="expenses", hue = "smoker", col = "sex", data = insurance, height =8,aspect=1)
plt.show()
```

```python
sns.lmplot(x = "bmi",y="expenses", hue = "smoker", col = "region",col_wrap=2, data = insurance, height =7,aspect=1)
plt.show()  #col_wrap=2 ikili ikili demek
```

```python
canada = pd.read_excel('data/canada1x.xlsx')
canada.head()
```

```python
canada.drop(columns = ['Type', 'Coverage', 'AREA', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName'])
```

```python
canada.rename(columns = {'OdName' : 'Country'}, inplace = True)
canada.set_index(canada.Country, inplace = True)
canada.drop(columns = ['Type', 'Coverage', 'AREA', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName'], axis = 1, inplace = True)
canada.head()
```

```python
canada.index.name = None
canada.head()
```

```python
del canada['Country']
canada.head()
```

```python
canada = canada.transpose()
canada.head()
```

```python
canada1 = canada.loc[2004:, ["Germany","Brazil","Israel","Jamaica","Argentina","Afghanistan","Zambia","Austria","Uruguay","Zimbabwe"]]
canada1.head()
                               
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1)
plt.yticks(rotation =0)
plt.show
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1, vmin = 0, vmax=4500) #veya center= 4000 ortasÄ±
plt.yticks(rotation =0)
plt.show
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1, center=4000) #veya center= 4000 ortasÄ±
plt.yticks(rotation =0)
plt.show
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1, vmin = 0, vmax=8000 ,cmap="coolwarm")
plt.yticks(rotation =0)
plt.show
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1, vmin = 0, vmax=8000 ,linewidth=0.1)
plt.yticks(rotation =0)
plt.show
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(canada1, vmin = 0, vmax=8000,cmap="YlGnBu" ,annot =True,linewidth=.5, fmt=".1f") #annot =True, fmt="d" deÄerleri verir
plt.yticks(rotation =0)
plt.show
```
