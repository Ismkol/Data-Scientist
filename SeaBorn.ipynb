```python
#area
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
mpl.rcParams.update(mpl.rcParamsDefault)

%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
plt.style.use("seaborn-darkgrid")
```

```python
col1 = np.linspace(0,10,100)
col2 = np.sin(col1)

df = pd.DataFrame({"C1":col1,"C2":col2})
df.head(10)
```

```python
sns.lineplot(x=df.C1,y=df.C2,data=df)
plt.show()
```

```python
plt.figure(figsize = (14,6))
sns.set(rc={"axes.facecolor" : "#fffde7","axes.grid": True, "xtick.labelsize":14,"ytick.labelsize":14})
sns.lineplot(x=df.C1,y=df.C2,  data=df,linewidth =2)
sns.lineplot(x=df.C1,y=df.C2+1,data=df,linewidth =2)
plt.show()
```

```python
plt.figure(figsize = (14,6))
sns.set(rc={"axes.facecolor" : "#283747","axes.grid": False, "xtick.labelsize":14,"ytick.labelsize":14})
sns.lineplot(x=df.C1, y=df.C2,         data=df, linewidth =2)
sns.lineplot(x=df.C1, y=np.cos(df.C1), data=df, linewidth =2)
plt.show()
```

```python
canada = pd.read_csv("data/4.csv", delimiter = ";")
canada.head(10)
```

```python
canada.columns
```

```python
canada.rename(columns = {'OdName' : 'Country'}, inplace = True)
canada.set_index(canada.Country, inplace = True)
canada.drop(columns = ['Type', 'Coverage', 'AREA', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName'], axis = 1, inplace = True)
canada.head()
```

```python
canada = canada.transpose()
canada.head()
```

```python
plt.figure(figsize = (16,6))
plt.title("Immigrants from Afghanistan",fontsize = 14)
sns.set(rc = {"axes.facecolor":"#283747", "axes.grid" : False, "xtick.labelsize":10, "ytick.labelsize":10}) 
plt.xticks(rotation =45)
sns.lineplot(x = canada.index.values, y= canada["Afghanistan"])
plt.show()
```

```python
plt.figure(figsize = (16,6))
plt.title("Immigrants from Turkey, Pakistan, India", fontsize = 14)
sns.set(rc = {"axes.facecolor" : "#283747", "axes.grid" : False, 'xtick.labelsize' : 10, 'ytick.labelsize' : 10})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['India'], color = '#ff9900', label = 'India')
sns.lineplot(x = canada.index.values, y = canada['Pakistan'], color = '#4586ff', label = 'Pakistan')
sns.lineplot(x = canada.index.values, y = canada['Turkey'], color = '#a2ef44', label = 'Turkey')
plt.legend(facecolor = 'grey', fontsize = 'large', edgecolor = 'black', shadow = True)
plt.show()
```

```python
employment = pd.read_csv("data/employment.csv", delimiter = ";")
employment.head(10)
```

```python
plt.figure(figsize = (16,6))
plt.style.use("seaborn-darkgrid")
sns.lineplot(x = "Period", y = "Unemployed", hue = "Gender", data = employment,linewidth = 2)
plt.show()
```

```python
plt.figure(figsize = (16,6))
plt.style.use("seaborn-darkgrid")
sns.lineplot(x = "Period", y = "Unemployed", hue = "Gender",markers = True, data = employment)
plt.show()
```

```python
plt.figure(figsize = (16,6))
plt.style.use("seaborn-darkgrid")
sns.lineplot(x = "Period", y = "Unemployed", hue = "Age",style = "Gender",estimator=None ,units="Age",markers = True, data = employment)
plt.legend(bbox_to_anchor = (1.0,1.0), shadow =True)
plt.show()
```
