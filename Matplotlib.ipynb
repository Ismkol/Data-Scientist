```python
#Multiple data representation
```

```python
plt.figure(figsize = (10,5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x,2)
y2 = np.power(x,3)
plt.plot(x,y1, "k-",label= "$y1 = x^2$")
plt.plot(x,y2, "r-",label= "$y2 = x^3$")
plt.xlabel("X-Eksen")
plt.ylabel("Y-Eksen")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0,10,1000)
plt.figure(figsize=(10,6))
plt.plot(x, np.sin(x), label = "$Sin(x)$")
plt.plot(x, np.cos(x), label = "$Cos(x)$")
plt.xlabel(r"$X$",fontsize =18)
plt.ylabel(r"$Y$",fontsize =18)
plt.title("$Sin(X) $ $&$ $Cos(X)$", fontsize = 14)
plt.legend(loc = "upper right")
plt.show()
```

```python
#Alan boyama (Shading ragiosn)
```

```python
x = np.linspace(0,10,1000)
plt.figure(figsize = (10,6))
plt.plot(x, np.sin(x),"k-", label ="$Sin(x)$")
plt.plot(x, np.cos(x), label ="$Cos(x)$")
plt.fill_between(x, 0, np.sin(x))
plt.fill_between(x, 0, np.cos(x))
plt.xlabel(r"$X$", fontsize = 18)
plt.ylabel(r"$Y$", fontsize = 18)
plt.title("$Sin(X) $ $&$ $Cos(X)$", fontsize = 20)
plt.legend(loc = "lower left")
plt.show()
```

```python
plt.figure(figsize = (10,6))
x = [1,2,3,4,5,6,7,8,9]
y = [7,8,9,10,7,11,12,13,14]
y2 = [7,7,7,7,7,7,7,7,7]
plt.plot(x,y,linewidth = 4, label ="Line-1")
plt.plot(x,y2,linewidth =3, label="Line-2")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line Graph")
plt.legend
plt.show
```

```python
plt.figure(figsize = (10,6))
x = [1,2,3,4,5,6,7,8,9]
y = [7,8,9,10,7,11,12,13,14]
y2 = [7,7,7,7,7,7,7,7,7]
plt.plot(x,y,linewidth = 4, label ="Line-1")
plt.plot(x,y2,linewidth =3, label="Line-2")
plt.xlabel("X",fontsize =20)
plt.ylabel("Y")
plt.title("Line Graph")
plt.legend()
plt.grid(b =True, linestyle =":",which ="major", color = "blue")
plt.show
```

```python
plt.figure(figsize = (10,5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x,2)
y2 = np.power(x,3)
ax = plt.axes()
ax.set_facecolor("darkgrey")
plt.plot(x,y1,"bo-", x,y2,"ro-")
plt.grid(b =True, linestyle ="-",which ="major", color = "black")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
