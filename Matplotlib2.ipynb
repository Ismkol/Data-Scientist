```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
plt.style.use("seaborn-darkgrid")
```

```python
#Coklu Grafik (Multiple Plots)
```

```python
plt.figure(figsize = (14,6))
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2= np.cos(x)
plt.subplot(1, 2, 1) #1 sat覺rl覺 2 sutunlu 1.grafik
plt.plot(x,y1, "r-") 
plt.subplot(1,2,2) #1 sat覺rl覺 2 sutunlu 2.grafik
plt.plot(x,y2, "b-")
plt.show()
```

```python
plt.figure(figsize = (14,6))
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2= np.cos(x)
plt.subplot(2, 1, 1) #2 sat覺rl覺 1 sutunlu 1.grafik
plt.plot(x,y1, "r-")
plt.grid(b =True, linestyle ="-",which ="major", color = "black")
plt.subplot(2,1,2)  #2 sat覺rl覺 1 sutunlu 2.grafik
plt.plot(x,y2, "b-")
plt.grid(b =True, linestyle =":",which ="major", color = "blue")
plt.show()
```

```python
x = np.arange(-50,50)
y1 = np.power(x,2)
y2 = np.power(x,3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)

fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize=(20,20))

ax1[0,0].plot(x,y1, "tab:blue")
ax1[0,0].set_title("Square Func")
ax1[0,0].set_xlabel(r"$X$", fontsize = 18)
ax1[0,0].set_ylabel(r"$Y$", fontsize = 18)

ax1[0,1].plot(x,y2, "tab:red")
ax1[0,1].set_title("Square Func")
ax1[0,1].set_xlabel(r"$X$", fontsize = 18)
ax1[0,1].set_ylabel(r"$Y$", fontsize = 18)

ax1[0,2].plot(x,y3, "tab:olive")
ax1[0,2].set_title("Sine Func")
ax1[0,2].set_xlabel(r"$X$", fontsize = 18)
ax1[0,2].set_ylabel(r"$Y$", fontsize = 18)

ax1[1,0].plot(x,y4, "tab:cyan")
ax1[1,0].set_title("Cosine Func")
ax1[1,0].set_xlabel(r"$X$", fontsize = 18)
ax1[1,0].set_ylabel(r"$Y$", fontsize = 18)

ax1[1,1].plot(x,y5, "g-")
ax1[1,1].set_title("Tan Func")
ax1[1,1].set_xlabel(r"$X$", fontsize = 18)
ax1[1,1].set_ylabel(r"$Y$", fontsize = 18)

ax1[1,2].plot(x,y6, "m-")
ax1[1,2].set_title("Hyperbolic Func")
ax1[1,2].set_xlabel(r"$X$", fontsize = 18)
ax1[1,2].set_ylabel(r"$Y$", fontsize = 18)

ax1[2,0].plot(x,y7, "k-")
ax1[2,0].set_title("Hyperbolic Func")
ax1[2,0].set_xlabel(r"$X$", fontsize = 18)
ax1[2,0].set_ylabel(r"$Y$", fontsize = 18)

ax1[2,1].plot(x,y8, "y-")
ax1[2,1].set_title("Hyperbolic Func")
ax1[2,1].set_xlabel(r"$X$", fontsize = 18)
ax1[2,1].set_ylabel(r"$Y$", fontsize = 18)

ax1[2,2].plot(x,y9, "tab:blue")
ax1[2,2].set_title("Exponantial Func")
ax1[2,2].set_xlabel(r"$X$", fontsize = 18)
ax1[2,2].set_ylabel(r"$Y$", fontsize = 18)

plt.show()
```

```python
#Grouped Bar Chart
```

```python
plt.figure(figsize = (7,9))
barWidth = 0.20

y1 = np.array([17,18,29,40])
y2 = np.array([20,21,22,23])
y3 = np.array([5,9,11,12])

pos1 = np.arange(len(y1))
pos2 = [x + barWidth for x in pos1]
pos3 = [x + barWidth for x in pos2]

plt.bar(pos1, y1, color="#FBC02D", width=barWidth, label="Sat覺a A癟覺k")
plt.bar(pos2, y2, color="#F57F17", width=barWidth, label="Sat覺a Kapal覺")
plt.bar(pos3, y3, color="yellowgreen", width=barWidth, label="Sat覺覺 襤ptal Edilmi")

plt.legend()
plt.show()
```

```python
x1 = ["Bur癟in","Esra","襤lke","襤smail"]
y1 = [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize = (8,3))
plt.barh(x1,y1, label = "Sat覺a A癟覺k Biletler", color = "#FF6F00")
plt.barh(x1,y2, label = "Sat覺覺 Tamamlanm覺 Biletler", left = y1 ,color = "#FFB300")
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.title("Bar Chart")
plt.legend(loc="lower right")
plt.show()
```

```python
x1 = ["Bur癟in","Esra","襤lke","襤smail"]
y1 = [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize = (8,3))
plt.barh(x1,y1, label = "Sat覺a A癟覺k Biletler", color = "#FF6F00")
plt.barh(x1,y2, label = "Sat覺覺 Tamamlanm覺 Biletler", left = y1 ,color = "#FFB300")
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.title("Bar Chart")

for x, y in zip(x1,y1):
    plt.text(y-10,x, "%d" %y, ha ="center", va = "bottom", fontsize =10)
    
for x, y, z in zip(x1,y2,y1):
    plt.text(y+z-10, x, "%d" %y, ha ="center", va = "bottom", fontsize =10)    

plt.legend(loc="lower right")
plt.show()
```

```python
plt.figure(figsize = (7,9))
barWidth = 0.20

y1 = np.array([17,18,29,40])
y2 = np.array([20,21,22,23])
y3 = np.array([5,9,11,12])

pos1 = np.arange(len(y1))
pos2 = [x + barWidth for x in pos1]
pos3 = [x + barWidth for x in pos2]

plt.bar(pos1, y1, color="#FBC02D", width=barWidth, label="Sat覺a A癟覺k")
plt.bar(pos2, y2, color="#F57F17", width=barWidth, label="Sat覺a Kapal覺")
plt.bar(pos3, y3, color="yellowgreen", width=barWidth, label="Sat覺覺 襤ptal Edilmi")

for x, y in zip(pos1,y1):
    plt.text(x,y, "%d" %y, ha ="center", va = "bottom", fontsize =10)
    
for x, y in zip(pos2,y2):
    plt.text(x,y, "%d" %y, ha ="center", va = "bottom", fontsize =10)
    
for x, y in zip(pos3,y3):
    plt.text(x,y, "%d" %y, ha ="center", va = "bottom", fontsize =10)
    
plt.legend()
plt.show()
```
