```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)

%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
plt.style.use("seaborn-darkgrid")
```

```python
#Pie Charts (Pasta Grafikleri)
```

```python
plt.figure(figsize = (9,9))
area= [48,30,20,15]

labels = ["Low","Medium","High","Critical"]
colors = ["#8BC34A","#D4E157","#FFB300","#FF7043"]
plt.pie(area, labels =labels, colors = colors, startangle = 45)

plt.show()
```

```python
plt.figure(figsize = (9,9))
area= [48,30,20,15]

labels = ["Low","Medium","High","Critical"]
colors = ["#8BC34A","#D4E157","#FFB300","#FF7043"]
plt.pie(area, labels =labels, colors = colors, startangle = 45,shadow = "true", autopct = "%1.1f%%")

plt.show()
```

```python
plt.figure(figsize = (9,9))
area= [48,30,20,15]

total= np.sum(area)

labels = ["Low","Medium","High","Critical"]

def val_per(x):
    return "{:.2f}%\n({:.1f})".format(x, total*x/100)

colors = ["#8BC34A","#D4E157","#FFB300","#FF7043"]
plt.pie(area, labels =labels, colors = colors, startangle = 45,shadow = "true", autopct = val_per)

plt.show()
```

```python
plt.figure(figsize = (9,9))
area= [48,30,20,15]

labels = ["Low","Medium","High","Critical"]
colors = ["#8BC34A","#D4E157","#FFB300","#FF7043"]
plt.pie(area, labels =labels, colors = colors, startangle = 45,shadow = "true", autopct = "%1.1f%%", explode = [0, 0, 0.1, 0.1])
plt.legend()

plt.show()
```

```python
plt.figure(figsize = (9,9))
area= [48,30,20,15]

labels = ["Low","Medium","High","Critical"]
colors = ["#8BC34A","#D4E157","#FFB300","#FF7043"]
plt.pie(area, labels =labels, colors = colors, startangle = 45)

my_circle = plt.Circle((0,0),0.7, color ='white')
fig = plt.gcf()
fig.gca().add_artist(my_circle)

plt.show()
```

```python
plt.figure(figsize = (9,9))
fig.patch.set_facecolor("#DADADA")
area= [48,30,20,15]

labels = ["Low","Medium","High","Critical"]
colors = ["#8BC34A","#D4E157","#FFB300","#FF7043"]
plt.pie(area, labels =labels, colors = colors, startangle = 45,explode = [0.03, 0.03, 0.03, 0.01])

my_circle = plt.Circle((0,0),0.7, color ='white')
fig = plt.gcf()
fig.gca().add_artist(my_circle)

plt.show()
```

```python
plt.figure(figsize = (9,9))
fig.patch.set_facecolor("#DADADA")
area= [48,30,20,15]

labels = ["Low","Medium","High","Critical"]
colors = ["#8BC34A","#D4E157","#FFB300","#FF7043"]
plt.pie(area, labels =labels, colors = colors, startangle = 45,explode = [0.05, 0.05, 0.05, 0.05], autopct='%1.1f%%', pctdistance = 0.8) #distance yüzdeleri ortaya getirdi

my_circle = plt.Circle((0,0),0.7, color ='white')
fig = plt.gcf()
fig.gca().add_artist(my_circle)

plt.show()
```

```python
plt.figure(figsize = (9,9))
area= [40,30,20,15]

total= np.sum(area)

labels = ["Low","Medium","High","Critical"]

def val_per(x):
    return "{:.2f}%\n({:.1f})".format(x, total*x/100)

colors = ["#8BC34A","#D4E157","#FFB300","#FF7043"]
plt.pie(area, labels =labels, colors = colors, startangle = 45,pctdistance = 0.85 , autopct = val_per)

my_circle = plt.Circle((0,0),0.7, color ='white')
fig = plt.gcf()
fig.gca().add_artist(my_circle)

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (20,9))

area= [48,30,20,15]

priority = ["Low","Medium","High","Critical"]
status = ["Resolved", "Cancelled", "Pending","Assigned"]
company = ["IBM","Tesla","Microsoft","Apple"]

colors = ["#8BC34A","#D4E157","#FFB300","#FF7043"]

plt.subplot(1, 3, 1)

plt.pie(area , labels = priority, colors=colors, startangle =45)
my_circle = plt.Circle((0,0),0.7, color ='white')
fig = plt.gcf()
fig.gca().add_artist(my_circle)

plt.subplot(1, 3, 2)
plt.pie(area , labels = status, colors=colors, startangle =45)
my_circle = plt.Circle((0,0),0.7, color ='white')
fig = plt.gcf()
fig.gca().add_artist(my_circle)

plt.subplot(1, 3, 3)
plt.pie(area , labels = company, colors=colors, startangle =45)
my_circle = plt.Circle((0,0),0.7, color ='white')
fig = plt.gcf()
fig.gca().add_artist(my_circle)

plt.show()
```
