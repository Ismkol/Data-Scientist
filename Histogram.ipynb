```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)

%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
plt.style.use("seaborn-darkgrid")
```

```python
x = np.random.normal(10,5,1000)
plt.hist(x)
plt.show()
```

```python
x = np.random.normal(size = 2000)
plt.hist(x, bins=40,color="yellowgreen")
plt.gca().set(title="Histogram",ylabel ="Frekans")
plt.show()
```

```python
x= np.random.rand(2000)
plt.hist(x,bins=30,color="#D4AC0D",alpha=0.8)
plt.gca().set(title="Histogram",ylabel ="Frekans")
plt.show()
```

```python
plt.figure(figsize=(10,8))
x = np.random.normal(size = 2000)
plt.hist(x, bins=20,color="yellowgreen",edgecolor="#6A9662")
plt.gca().set(title="Histogram",ylabel ="Frekans")
plt.show()
```

```python
plt.figure(figsize = (8,11))
x = np.random.normal(-4,1,  size=800)
y = np.random.normal(0,1.5, size=800)
z = np.random.normal(3.5,1, size=800)

plt.hist(x, bins=30,color="yellowgreen",alpha =0.5)
plt.hist(y, bins=30,color="#6A9662",alpha =0.5)
plt.hist(z, bins=30,color="#FFB300",alpha =0.5)

plt.gca().set(title="Histogram",ylabel ="Frekans")
plt.show()
```

```python
#Cumulative Distribution Function (Artarak dağılarn fonk)
```

```python
plt.figure(figsize = (10,8))
x= np.random.rand(2000)
plt.hist(x,bins=30,color="#ffa41b",edgecolor="#639a67",cumulative = True)
plt.gca().set(title="Histogram",ylabel ="Frekans")
plt.show()
```

```python
#AREA PLOT
```

```python
x = np.arange(1,31)
y = np.random.normal(10,11,size =30)
y= np.square(y)

plt.figure(figsize =(16,6))
plt.plot(x,y, color ="#baf1a1")
plt.fill_between(x,y, color ="#7fcd91")
plt.show()
```
